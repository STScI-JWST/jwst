### MIRI

Let's get the example pool

```python
pool_miri = AssociationPool.from_file('../data/jw93060_001_01_pool.fits')
```

```python
pool_miri.pool_params
```

```python
# Use warnings for processing mishaps that do not halt execution.
# Show all instances.
with warnings.catch_warnings():
    #warnings.simplefilter('ignore', RuntimeWarning)
    (assocs_miri, orphaned) = pool_miri.generate(assoc_definitions)
```

```python
print 'There where {:d} associations found.'.format(len(assocs_miri))
print type(assocs_miri[0])
for assocs in assocs_miri:
    print assocs
```

```python
print '\n There were {} orphaned entries.'.format(len(orphaned))
```

```python
assocs_miri[0].format('hdulist')
```

### Check for JSON'ness

```python
assocs_miri[0].to_json()
```

#### Convert the HDUlists

```python
for assocs in assocs_miri:
    (asn_name, asn_hdulist) = assocs.to_hdulist()
    print '\nProduct {}'.format(asn_name)
    print repr(asn_hdulist[0].header)
    print asn_hdulist[1].columns.names
    print asn_hdulist[1].data
```
