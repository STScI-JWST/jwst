## Environment Setup

```python
import subprocess
import os
env=os.environ.copy()
env['PYTHONPATH'] = '../../jwst_tools/associations/'
```

```python
def runcmd(cmd):
    try:
        print subprocess.check_output(cmd.split(' '), env=env, stderr=subprocess.STDOUT)
    
    except subprocess.CalledProcessError as e:
        print '******** Error! Output as follows:\n'
        print e.output
```

## Examples

```python
cmd='../../scripts/asn_generate ../data/jw93060_001_01_pool.csv ../data/association_definitions.py'
runcmd(cmd)
```
