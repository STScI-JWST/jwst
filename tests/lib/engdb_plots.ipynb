# Library

```python
def engplot(mnemonic, stime='2016-01-01', etime='2016-12-31'):
    values = edb.get_values(mnemonic, stime, etime, include_obstime=True, zip=False)
    dt = list(pd.to_datetime([o.unix for o in values.obstime], unit='s'))
    df = {
        'obstime': dt,
        'value': values.value
    }
    p = TimeSeries(df, x='obstime', y='value', builder_type='line', title=mnemonic)
    return p
```

```python
def engplot_update(mnemonic, stime='2016-01-22', etime='2016-12-23'):
    try:
        p = engplot(mnemonic=mnemonic, stime=stime, etime=etime)
    except:
        pass
    else:
        r.children[0] = p
        push_notebook(handle=handle)
```

Get the list of mnemonics

```python
p = engplot('SA_ZATTEST1')
r = row(p)
handle = show(r, notebook_handle=True)
```

```python
interact(
    engplot_update, 
    mnemonic=mnemonics,
    stime='2016-01-01',
    etime='2016-12-31',
    __manual=True
)
```
